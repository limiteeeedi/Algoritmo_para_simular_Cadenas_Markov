```python
import random

# Elegir estado según probabilidades
def elegir(p):
    r = random.random()
    s = 0
    for i in range(len(p)):
        s += p[i]
        if r <= s:
            return i

# Generar salida
def simCM(a, P, n):
    X = [elegir(a)]
    for i in range(1, n+1):
        j = X[i-1]
        p = P[j]
        X.append(elegir(p))
    return X

# ENTRADAS
colores = ["Naranja", "Azul", "Verde", "Rojo"]   # estados
a = [0.25, 0.25, 0.25, 0.25]   # distribucion inicial
P = [
    [1/4, 0,   1/4, 2/4],
    [2/4, 1/4, 0,   1/4],
    [1/4, 2/4, 1/4, 0  ],
    [0,   1/4, 2/4, 1/4]
]   # matriz de transicion
n = 100   # numero de pasos

#SALIDA
resultado = simCM(a, P, n)
print([colores[i] for i in resultado])
```

Tenemos un tablero de 4 sectores. Lanzamos 2 monedas y avanzamos en sentido horario el numero de caras que aparezcan.

La distribución inicial es uniforme para los 4 estados posibles (naranja, azul, verde, rojo).
